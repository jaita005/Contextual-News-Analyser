```python
from newspaper import Article
from textblob import TextBlob
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lsa import LsaSummarizer
import nltk

# Download 'punkt' tokenizer
nltk.download('punkt')

# Step 1: Load the article
url = 'https://edition.cnn.com/2020/09/13/tech/microsoft-tiktok-bytedance/index.html'
article = Article(url)
article.download()
article.parse()

# Step 2: Print the article title and text
print("üì∞ Title:", article.title)
print("\nüìÑ Text Snippet:", article.text[:500], "...")

# Step 3: Sentiment analysis using TextBlob
blob = TextBlob(article.text)
print("\nüîç Sentiment:")
print("   Polarity:", blob.sentiment.polarity)
print("   Subjectivity:", blob.sentiment.subjectivity)

# Step 4: Summarize using sumy (3-sentence summary)
parser = PlaintextParser.from_string(article.text, Tokenizer("english"))
summarizer = LsaSummarizer()
summary = summarizer(parser.document, 3)

print("\nüìå Summary:")
for sentence in summary:
    print("-", sentence)
```

```python
from newspaper import Article
from textblob import TextBlob
from gensim.summarization import summarize
import nltk

# Download punkt tokenizer (already done, but do again just in case)
nltk.download('punkt')

# Step 1: Load the article
url = 'https://edition.cnn.com/2020/09/13/tech/microsoft-tiktok-bytedance/index.html'
article = Article(url)
article.download()
article.parse()

# Step 2: Print title and sample of article text
print("üì∞ Title:", article.title)
print("\nüìÑ Text Snippet:", article.text[:500], "...")

# Step 3: Sentiment analysis using TextBlob
blob = TextBlob(article.text)
print("\nüîç Sentiment:")
print("   Polarity:", blob.sentiment.polarity)
print("   Subjectivity:", blob.sentiment.subjectivity)

# Step 4: Summarize using gensim
try:
    summary = summarize(article.text, word_count=100)  # you can also use ratio=0.1
    print("\nüìå Summary:\n", summary)
except ValueError:
    print("\n‚ö†Ô∏è Article is too short to summarize using gensim.")
```

```python
from newspaper import Article
from textblob import TextBlob
from summa.summarizer import summarize
from transformers import pipeline
import nltk

# Download necessary tokenizer
nltk.download('punkt')

# Load article
url = 'https://edition.cnn.com/2020/09/13/tech/microsoft-tiktok-bytedance/index.html'
article = Article(url)
article.download()
article.parse()

# Print basic details
print("üì∞ Title:", article.title)
print("‚úçÔ∏è Authors:", article.authors)
print("\nüìÑ Text Snippet:", article.text[:500], "...")

# Sentiment analysis
blob = TextBlob(article.text)
print("\nüîç Sentiment:")
print("   Polarity:", blob.sentiment.polarity)
print("   Subjectivity:", blob.sentiment.subjectivity)

# Summary using summa
summary = summarize(article.text, words=100)
print("\nüìå Summary:\n", summary)

# Topic Classification using Hugging Face's zero-shot classifier
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# You can change these labels based on your use case
candidate_labels = ["Politics", "Technology", "Business", "Health", "Sports", "Entertainment"]

classification = classifier(article.text[:1000], candidate_labels)
print("\nüóÇÔ∏è Category Prediction:")
print("   Predicted:", classification['labels'][0])
print("   All Scores:", dict(zip(classification['labels'], classification['scores'])))
```

```python
import tkinter as tk
from newspaper import Article
from textblob import TextBlob
from summa.summarizer import summarize
from transformers import pipeline
import nltk

classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
candidate_labels = ["Politics", "Technology", "Health", "Sports", "Entertainment", "Business", "Science"]

def summarize_article():
    url = utext.get('1.0', "end").strip()

    article = Article(url)
    article.download()
    article.parse()

    title.config(state="normal")
    author.config(state="normal")
    publication.config(state="normal")
    summary.config(state="normal")
    sentiment.config(state="normal")
    category.config(state="normal")

    title.delete('1.0', 'end')
    title.insert('1.0', article.title)

    author.delete('1.0', 'end')
    author.insert('1.0', ', '.join(article.authors))

    publication.delete('1.0', 'end')
    publication.insert('1.0', str(article.publish_date))

    blob = TextBlob(article.text)
    sentiment.delete('1.0', 'end')
    sentiment.insert('1.0', f'Polarity: {blob.sentiment.polarity}\tSubjectivity: {blob.sentiment.subjectivity}\nüîç Sentiment: {"positive" if blob.sentiment.polarity > 0 else "negative" if blob.sentiment.polarity < 0 else "neutral"}')

    result = classifier(article.text[:1000], candidate_labels)

    category.delete('1.0', 'end')
    category.insert('1.0', f'Predicted Category: {result["labels"][0]} ({result["scores"][0]:.2f})')
    
    summary_text = summarize(article.text, words=100)
    summary.delete('1.0', 'end')
    summary.insert('1.0', summary_text if summary_text else "Summary not available for this article.")

    title.config(state="disabled")
    author.config(state="disabled")
    publication.config(state="disabled")
    summary.config(state="disabled")
    sentiment.config(state="disabled")
    category.config(state="disabled")

root = tk.Tk()
root.title("News Summarizer")
root.geometry('1200x900')

tk.Label(root, text="Title").pack()
title = tk.Text(root, height=1, width=140, bg="#dddddd", state="disabled")
title.pack()

tk.Label(root, text="Authors").pack()
author = tk.Text(root, height=1, width=140, bg="#dddddd", state="disabled")
author.pack()

tk.Label(root, text="Publishing Date").pack()
publication = tk.Text(root, height=1, width=140, bg="#dddddd", state="disabled")
publication.pack()

tk.Label(root, text="Summary").pack()
summary = tk.Text(root, height=20, width=140, bg="#dddddd", state="disabled")
summary.pack()

tk.Label(root, text="Sentiment Analysis").pack()
sentiment = tk.Text(root, height=2, width=140, bg="#dddddd", state="disabled")
sentiment.pack()

tk.Label(root, text="Category Analysis").pack()
category = tk.Text(root, height=1, width=140, bg="#dddddd", state="disabled")
category.pack()

tk.Label(root, text="URL").pack()
utext = tk.Text(root, height=1, width=140)
utext.pack()

tk.Button(root, text="Summarize", command=summarize_article).pack()

root.mainloop()
```
